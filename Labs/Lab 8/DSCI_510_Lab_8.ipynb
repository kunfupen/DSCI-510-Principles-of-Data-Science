# 1. Basic Socket Request


```python
import socket

my_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

my_socket.connect(('example.com', 80))

request = "GET / HTTP/1.1\r\nHost: example.com\r\n\r\n"

my_socket.sendall(request.encode())

response = my_socket.recv(1024)

print(response.decode('utf-8'))

my_socket.close()
```

# 2. Bytes and Strings Conversion

```python
input_string = input("Enter a string: ")

byte_data = input_string.encode('utf-8')
print(byte_data)

decoded_string = byte_data.decode('utf-8')
print(decoded_string)
```

# 3. HTTP Request Inspection


```python
import requests
url = 'http://universities.hipolabs.com/search?name=california'
r = requests.get(url)

print(f"Status Code: {r.status_code}")
print(r.headers)
print(f"Final URL: {r.url}")

universities = r.json()

print(f"Number of universities returned: {len(universities)}")

for i, university in enumerate(universities[:3], 1):
    print(f"name: {university['name']}, country: {university['country']}, web_page: {university['web_pages'][0]}")


```
