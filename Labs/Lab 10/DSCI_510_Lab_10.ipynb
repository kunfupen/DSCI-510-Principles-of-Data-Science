# Skewness

```python
import pandas as pd
import numpy as np

def skewness(x):
    n = len(x)
    mean_x = np.mean(x)
    std_x = np.std(x, ddof=1)
    skew =  1/ n * np.sum(((x - mean_x) / std_x) ** 3)
    return skew

arr = np.array([1, 2, 3, 4, 5, 5])
print(skewness(arr))
```

# Fetch JSON Data and Convert to CSV

```python
import requests
import pandas as pd

def fetch_json_data():
    url = "https://jsonplaceholder.typicode.com/users"

    response = requests.get(url)
    data = response.json()

    extracted_data = []
    for user in data:
        extracted_data.append({
            "id": user["id"],
            "name": user["name"],
            "email": user["email"],
            "city": user["address"]["city"]
        })

    df = pd.DataFrame(extracted_data)

    df.to_csv("users_data.csv", index=False)

fetch_json_data()
```
