```python
import json
import numpy as np
import pandas as pd
```

```python
a = np.array([1., 4., 2., 12.3], dtype=np.float32)

# sum 
print(np.sum(a))
print(a.sum())

# mean
print(np.mean(a))

# std
print(np.std(a))
```

```python
print(a[:3]) # slicing
print(a[::-1])
```

```python
b = np.array([3, 5, 8, 1], dtype=np.float32)
print(a)
print(b)
print(a * b)  # element-wise multiplication

# adding
print(a + b)

# division
print(a / b)
```

```python
# 2D matrix
M = np.random.rand(3, 4)  # 3 rows, 4 columns
print(M)

print(M[:3, :2])  # first 3 rows, first 2 columns
```

```python
a = np.random.rand(4, 2) # 4 rows, 2 columns

print(M @ a) # matrix multiplication
print(np.dot(M, a))
```

```python
# compute the sum of each row
print(np.sum(M, axis=1)) # [3, ]

# compute the sum of each column
print(np.sum(M, axis=0)) # [4, ]
```

```python
data_dict = [
  {
    "id": 1,
    "name": "Leanne Graham",
    "username": "Bret",
    "email": "Sincere@april.biz",
    "address": {
      "street": "Kulas Light",
      "suite": "Apt. 556",
      "city": "Gwenborough",
      "zipcode": "92998-3874",
      "geo": {
        "lat": "-37.3159",
        "lng": "81.1496"
      }
    },
    "phone": "1-770-736-8031 x56442",
    "website": "hildegard.org",
    "company": {
      "name": "Romaguera-Crona",
      "catchPhrase": "Multi-layered client-server neural-net",
      "bs": "harness real-time e-markets"
    }
  }
]

with open('data.json', 'w') as f:
    json.dump(data_dict, f, indent=2)
```

```python
# load data from file
with open('data.json', 'r') as f:
    data = json.load(f)

print(data)
```

```python
print(type(data))  # list
print(type(data[0]))  # dict
```

```python
print(data[0]["name"])
```

```python
# convert to DataFrame
df = pd.read_json('data.json')
# print(df)
print(df.columns)
```

```python
df2 = pd.DataFrame(data)
print(df2.columns)
```

```python
# method 1: convert to dataframe directly, and select columns
df_selected = df2[['id', 'name', 'email', 'address']]
# print(df_selected)
# Need to process the 'address' column to extract the city

for i, row in df_selected.iterrows():
    address = row['address']
    city = address['city']
    df_selected.at[i, 'city'] = city
print(df_selected)
```

```python
# method 2: create new pd.DataFrame and loop through data
df = pd.DataFrame(columns=['id', 'name', 'email', 'city'])
print(df)
for i, item in enumerate(data):
    id_ = item['id']
    name = item['name']
    email = item['email']
    city = item['address']['city']
    df.loc[i] = [id_, name, email, city]
print(df)
```

```python
# select id == 1 and name == 'Leanne Graham'
print(df[(df['id'] == 1) & (df['name'] == 'Leanne Graham')]) # & is and operator, | is or operator

# select id > 1
print(df[df['id'] > 1])
```

```python
# save to csv
file_name = 'output.csv'
df.to_csv(file_name, index=False)
```
