# Prime Number Finder

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False

    return True
while True:
    try:
        num = int(input("Enter a positive integer: "))
        if num <= 0:
            print("Please enter a positive integer.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a positive integer.")

print(f"Prime numbers up to {num}:")
for i in range(2, num + 1):
    if is_prime(i):
        print(i, end=' ')
```

# CD Profit Calculator

```python
def compute_profit(initial_deposit, months):
    annual_rate = 0.04
    monthly_rate = annual_rate / 12

    total_interest = initial_deposit * monthly_rate * months
    curr_bal = initial_deposit + total_interest

    if months % 12 != 0:
        penalty = curr_bal * monthly_rate * 3
        final_balance = curr_bal - penalty
    else:
        final_balance = curr_bal

    total_profit = final_balance - initial_deposit
    return total_profit

initial = float(input("Enter initial deposit: "))
mnths = int(input("Enter number of months until withdrawal: "))

profit = compute_profit(initial, mnths)
print(f"Profit: ${profit:.2f}")
```

# 2-D Bag Managemenent System

```python
def add_item(bag, name, row, col, height, width):
    if row < 0 or col < 0 or row + height > len(bag) or col + width > len(bag[0]):
        print("Item does not fit in the bag.")
        return False

    for r in range(row, row + height):
        for c in range(col, col + width):
            if bag[r][c] != '.':
                print("Error: Item overlaps with existing item.")
                return False
    for r in range(row, row + height):
        for c in range(col, col + width):
            bag[r][c] = name[0].upper()
            
    print(f"Item {name} added")
def move_item(bag, name, new_row, new_col):
    item_pos = []
    for r in range(len(bag)):
        for c in range(len(bag[0])):
            if bag[r][c] == name[0].upper():
                item_pos.append((r, c))
    if not item_pos:
        print(f"Error: Item {name} not found.")
        return False
    

    min_r = min(pos[0] for pos in item_pos)
    max_r = max(pos[0] for pos in item_pos)
    min_c = min(pos[1] for pos in item_pos)
    max_c = max(pos[1] for pos in item_pos)

    height = max_r - min_r + 1
    width = max_c - min_c + 1

    if new_row < 0 or new_col < 0 or new_row + height > len(bag) or new_col + width > len(bag[0]):
        print("Error: New position out of bounds.")
        return False
    
    for r in range(new_row, new_row + height):
        for c in range(new_col, new_col + width):
            if bag[r][c] != '.' and bag[r][c] != name[0].upper():
                print("Error: New position overlaps with existing item.")
                return False
            
    for r, c in item_pos:
        bag[r][c] = '.'

    for r in range(new_row, new_row + height):
        for c in range(new_col, new_col + width):
            bag[r][c] = name[0].upper()
    print(f"Item {name} moved")
    return True
def print_bag(bag):
    for row in bag:
        print(' '.join(row))

row = int(input("Enter bag rows: "))
col = int(input("Enter bag columns: "))
bag = [['.' for _ in range(col)] for _ in range(row)]

command = input("Enter command (add/move/print/quit): ").strip().lower()
while command != 'quit':
    if command == 'add':
        name = input("Enter item name: ")
        top_left_row = int(input("Top-left row: "))
        top_left_col = int(input("Top-left column: "))
        height = int(input("Item height: "))
        width = int(input("Item width: "))
        add_item(bag, name, top_left_row, top_left_col, height, width)
    elif command == 'move':
        name = input("Enter item name to move: ")
        new_r = int(input("Enter new row: "))
        new_c = int(input("Enter new column: "))
        move_item(bag, name, new_r, new_c)
    elif command == 'print':
        print_bag(bag)
    else:
        print("Invalid command.")
    command = input("Enter command (add/move/print/quit): ").strip().lower()
```
