# Email Matcher with Regex

```python
import re
def is_valid_email(email):

    pattern = r'^[\w.\-]+@([\w.\-]+\.[a-zA-Z]{2,6})$'
    match = re.match(pattern, email)
    if match:
        domain = match.group(1)
        return True, domain
    else:
        return False, None

print(is_valid_email("alice@cs.ucla.edu"))
print(is_valid_email("bad-email@@test"))
print(is_valid_email("user@domaincom.a.a"))
```

# Product Sales Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales.csv')

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
products = df['Product'].tolist()

plt.figure(figsize=(10, 6))
for product in products:
    sales_data = df[df['Product'] == product][months].values.flatten()
    plt.plot(months, sales_data, marker='o', label=product)
plt.title('Monthly Sales Data')
plt.xlabel('Month names (Jan to May)')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()

total_sales = df[months].sum().values
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=months, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution from Jan to May')
plt.show()

total_units_sold_per_product = df[months].sum(axis=1)
plt.figure(figsize=(6, 6))
plt.bar(products, total_units_sold_per_product, color='lightblue')
plt.title('Total Units Sold per Product')
plt.xlabel('Product names')
plt.ylabel('Total Units Sold')
plt.show()
```
